# Conclusions:-

1.Bank can decrease the interest on the loan amount for the customers whose income range is low

2.For the consumers having high credit has less payment diificulties for which bank wont need to bear losses and can approve their loans.

3.Occupation type such as labours and sales staff are not recommended for loans where as Accountants are recommended.

4.Bank can see the family status i.e families with more children will have high payment difficulties so can approve loans to unmarried or only couples.

5.Banks should focus more on contract type ‘Student’ ,’pensioner’ and ‘Businessman’ with housing ‘type other than ‘Co-op apartment’ for successful payments.

6.Banks should focus less on income type ‘Working’ as they are having most number of unsuccessful payments.
